El gr√°fico de barras nos permite graficar variables categ√≥ricas, es decir, variables de texto, lo que es muy importante en el mundo de la ciencia de datos y Matplotlib ofrece ciertas caracter√≠sticas que nos facilita la vida en cuanto a graficar este tipo de variables.

Vamos al c√≥digo para que sepas c√≥mo hacerlo
* Importamos librer√≠as


```python
import matplotlib.pyplot as plt
import numpy as np
```

* Variables categ√≥ricas


```python
countrys = ["INDIA", "JAPAN", "MEXICO", "COLOMBIA", 'GERMANY']
population = [5000, 800, 900, 1000, 300]
```

* Gr√°fica de barras


```python
plt.bar(countrys,population)
plt.show()
```

* Modificar colores y tama√±o


```python
plt.bar(countrys,population,  width=0.5, color= ["aqua", "grey", "teal", "crimson", "violet"])
plt.show()
```

* Modificar los nombres de las variables en el eje ‚ÄòX‚Äô con xticks


```python
plt.bar(countrys,population, width=0.5, color= ["aqua", "grey", "teal", "crimson", "violet"])
plt.xticks(np.arange(5), ('India','Japon', 'Mexico', 'Colombia', 'Alemania'), rotation = 45)
plt.show()
```

* Crear un gr√°fico de manera horizontal con barh


```python
plt.barh(countrys,population)
plt.show()
```
